## 1. Introdução

A **Visão Computacional** é a área que estuda como extrair informação útil de imagens e vídeos
de forma automática, reproduzindo computacionalmente parte da capacidade humana de "enxergar".
Suas etapas básicas envolvem **adquirir** a imagem (de um arquivo ou de uma câmera),
**processá-la** (filtros, conversões de cor, detecção de bordas, etc.) e **gerar uma saída**
(exibição em tela ou gravação em arquivo).

O **OpenCV** (*Open Source Computer Vision Library*) é a biblioteca mais utilizada para esse fim.
Neste laboratório usamos sua interface Python (`cv2`), que oferece funções diretas para as quatro
operações fundamentais de entrada/saída exploradas aqui:

| Leitura de imagem | arquivo `.jpg` | `cv.imread` / `cv.imshow` | janela / `cv.imwrite` |
| Leitura de vídeo | arquivo `.mp4` | `cv.VideoCapture` | janela quadro a quadro |
| Captura de webcam | câmera (índice 0) | `cv.VideoCapture(0)` | janela / foto `.png` |
| Gravação de vídeo | câmera (índice 0) | `cv.VideoWriter` | arquivo `.avi` |

## 3. Procedimentos experimentais (Parte 1)

### 3.1 (A) Leitura de imagem em arquivo

**Objetivo:** ler uma imagem gravada em arquivo (`messi5.jpg`), exibi-la em uma janela e permitir
salvá-la em outro arquivo.

**Fundamentação:** `cv.imread(caminho, flag)` carrega a imagem. O segundo argumento (*flag*)
controla **como** a imagem é decodificada:

| Valor | Constante | Efeito |
|-------|-----------|--------|
| `1` | `cv.IMREAD_COLOR` | carrega em **cores** (BGR), padrão |
| `0` | `cv.IMREAD_GRAYSCALE` | carrega em **tons de cinza** |
| `-1` | `cv.IMREAD_UNCHANGED` | carrega com canal alfa, se houver |


````python
import numpy as np
import cv2 as cv

# Objetivo: ler a imagem do arquivo em ESCALA DE CINZA e exibi-la.
# Entrada: arquivo 'messi5.jpg' (flag 0 = cv.IMREAD_GRAYSCALE).
# Processamento: nenhum além da decodificação em cinza.
# Saída: janela 'image'; opcionalmente grava 'messigray.png'.
img = cv.imread('messi5.jpg', 0)        # flag 0 -> carrega em tons de cinza
cv.imshow('image', img)

k = cv.waitKey(0)                        # aguarda indefinidamente por uma tecla
if k == 27:                              # tecla ESC -> apenas fecha a janela
    cv.destroyAllWindows()
elif k == ord('s'):                      # tecla 's' -> salva a imagem em cinza e fecha
    cv.imwrite('messigray.png', img)
    cv.destroyAllWindows()
````

**Pergunta do enunciado — "Por que a janela aberta não mostra a imagem colorida?"**

Porque a imagem é lida em **escala de cinza**. Na chamada `cv.imread('messi5.jpg', 0)`, o
segundo argumento `0` corresponde à *flag* `cv.IMREAD_GRAYSCALE`, que instrui o OpenCV a
**descartar a informação de cor** e decodificar apenas a luminância (um único canal). Logo, a
janela exibe a imagem em tons de cinza, e não colorida. Para exibi-la em cores bastaria trocar a
*flag* para `1` (`cv.IMREAD_COLOR`), isto é, `cv.imread('messi5.jpg', 1)`.

**Resultado/execução:** ao rodar `python3 L__1_img.py`, abre-se a janela com a foto do Messi em
cinza; pressionar **`s`** grava `messigray.png` e **`ESC`** fecha a janela.


### 3.2 (B) Leitura de vídeo em arquivo

**Objetivo:** ler um vídeo gravado (`big_buck_bunny.mp4`) e exibir a sequência de quadros em uma
janela; em seguida, **alterar a velocidade** de exibição (mais rápido e mais lento).

**Fundamentação:** `cv.VideoCapture(arquivo)` abre o vídeo; o laço lê quadro a quadro com
`cap.read()`, que retorna `ret` (booleano de sucesso) e `frame` (a imagem). Cada quadro é
mostrado com `cv.imshow`.


````python
import time
import numpy as np
import cv2 as cv

# Objetivo: ler o vídeo do arquivo e exibi-lo quadro a quadro.
# Entrada: arquivo 'big_buck_bunny.mp4' (24 fps).
# Processamento: nenhum; controle de VELOCIDADE pelo atraso entre quadros.
# Saída: janela 'frame' exibindo a sequência de imagens.
cap = cv.VideoCapture('big_buck_bunny.mp4')

while cap.isOpened():
    ret, frame = cap.read()              # ret=False quando o vídeo termina

    if ret == True:
        cv.imshow('frame', frame)        # exibe o quadro atual

        # Atraso adicional entre quadros. Valor pequeno (1 ms) -> reprodução RÁPIDA.
        time.sleep(1/1000.0)

        # waitKey(1): aguarda 1 ms e permite sair com a tecla 'q'.
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break                            # fim do vídeo

cap.release()
cv.destroyAllWindows()
````

**Pergunta do enunciado — explicação da alteração de velocidade e soluções da equipe.**

A velocidade de exibição **não** depende dos dados do vídeo, e sim do **tempo de espera entre a
exibição de um quadro e o próximo**. Quem cria esse tempo de espera é a função
**`cv.waitKey(delay)`**, que pausa o programa por `delay` milissegundos (e, de quebra, captura
teclas). Assim:

- **Como o OpenCV controla a velocidade:** o tempo total por quadro é, aproximadamente,
  `delay` (do `waitKey`) somado a eventuais `time.sleep` e ao próprio custo de processamento.
- **Papel do `waitKey()`:** além de capturar teclas, é ele que **dá o ritmo** da reprodução; sem
  uma espera, os quadros passam praticamente instantaneamente.
- **Relação entre atraso e FPS:** a taxa de exibição é aproximadamente
  `fps ≈ 1000 / atraso_em_ms`. Para reproduzir na velocidade *natural* de um vídeo de 24 fps, o
  atraso deve ser `1000/24 ≈ 42 ms`.
- **Como acelerar:** **diminuir** o atraso (atraso menor → mais quadros por segundo).
- **Como desacelerar:** **aumentar** o atraso (atraso maior → menos quadros por segundo).

**Soluções concretas usadas/testadas pela equipe** (o vídeo `big_buck_bunny.mp4` tem 24 fps):

| Efeito | Ajuste | Atraso por quadro | Taxa aprox. |
|--------|--------|-------------------|-------------|
| Velocidade **natural** | `cv.waitKey(42)` | ~42 ms | ~24 fps |
| **Rápida** (versão entregue) | `cv.waitKey(1)` + `time.sleep(1/1000.0)` | ~1 ms | muito > 24 fps |
| **Lenta** | `cv.waitKey(100)` | ~100 ms | ~10 fps |

> **Inconsistência corrigida:** o comentário original do código dizia
> *"wait next frame by 40ms - 25fps"*, mas a espera efetiva era `time.sleep(1/1000.0)` com
> `waitKey(1)`, que resulta em reprodução **muito rápida** (≈1 ms por quadro), e não 40 ms. Além
> disso, o uso de `time.sleep` é redundante: o controle de velocidade pode (e deve, por
> simplicidade) ser feito apenas pelo parâmetro de `cv.waitKey`. Para a velocidade natural do
> vídeo, o correto é `cv.waitKey(42)`.


### 3.3 (C) Captura de imagem da webcam

**Objetivo:** ler imagens da webcam e exibi-las; em seguida, **salvar um quadro em arquivo**
quando a tecla **`x`** for pressionada. O enunciado pede que o arquivo se chame **`foto1.png`**.

**Fundamentação:** `cv.VideoCapture(0)` abre a câmera padrão (índice 0). O laço lê quadros
continuamente; `cv.waitKey(1)` captura a tecla pressionada a cada iteração.


> **Inconsistência encontrada no script original (`L__3_webcam.py`):** ele salvava o arquivo
> como `pessoalento.png` (e não `foto1.png`, como pede o item C) e chamava `cv.waitKey(1)`
> **duas vezes** no mesmo laço — cada chamada consome um evento de teclado distinto, o que torna
> a detecção das teclas pouco confiável. A versão abaixo **corrige** ambos os pontos: lê a tecla
> **uma única vez** por iteração e grava em **`foto1.png`**, conforme o enunciado.


````python
import numpy as np
import cv2 as cv

# Objetivo: capturar imagens da webcam e salvar um quadro como 'foto1.png' ao pressionar 'x'.
# Entrada: webcam (cv.VideoCapture(0)).
# Processamento: nenhum; apenas captura e detecção de teclas.
# Saída: janela 'frame' ao vivo e arquivo 'foto1.png' quando 'x' é pressionado.
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

while True:
    ret, frame = cap.read()              # ret=False indica falha/fim do stream
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    cv.imshow('frame', frame)            # exibe o vídeo ao vivo da câmera

    key = cv.waitKey(1) & 0xFF           # lê a tecla UMA vez por iteração
    if key == ord('q'):                  # 'q' encerra o programa
        break
    elif key == ord('x'):                # 'x' salva o quadro atual em arquivo
        cv.imwrite('foto1.png', frame)
        print("Imagem salva em foto1.png")

cap.release()
cv.destroyAllWindows()
````

**Explicação detalhada (pergunta do enunciado):**

- **Como a tecla é capturada:** `cv.waitKey(1)` aguarda 1 ms por um evento de teclado e retorna o
  código da tecla. O `& 0xFF` mantém apenas os 8 bits relevantes (boa prática para portabilidade
  entre sistemas). Comparamos o resultado com `ord('x')`.
- **Como o quadro é armazenado:** cada `cap.read()` devolve o quadro corrente em `frame`, um array
  NumPy (`altura × largura × 3`, em BGR). No instante em que `x` é pressionada, esse `frame` em
  memória é exatamente a imagem que será gravada.
- **Como a imagem é salva:** `cv.imwrite('foto1.png', frame)` codifica o array e grava em disco; a
  extensão `.png` define automaticamente o formato.

**Resultado/execução:** ao rodar o programa, a janela mostra a câmera ao vivo; pressionar **`x`**
grava `foto1.png` e **`q`** encerra. A equipe utilizou esta lógica de captura para produzir as
fotos da Parte 2 (foto geral da equipe).


### 3.4 (D) Gravação de vídeo da webcam

**Objetivo:** ler imagens da webcam, exibi-las e **gravar** a sequência em arquivo `.avi`. O item
pede que as imagens gravadas fiquem **"normais"** (orientação correta) e com **velocidade de
exibição adequada**.

**Fundamentação:** `cv.VideoWriter(arquivo, fourcc, fps, (largura, altura))` cria o gravador. O
`fourcc` define o codec (`XVID` para AVI); `fps` define a taxa de quadros gravada; o tamanho deve
**coincidir** com o tamanho dos quadros capturados, senão o arquivo sai vazio/corrompido.


````python
import numpy as np
import cv2 as cv

# Objetivo: capturar da webcam e gravar a sequência em um arquivo .avi.
# Entrada: webcam (cv.VideoCapture(0)).
# Processamento: NENHUM flip -> a imagem é gravada na orientação NORMAL.
# Saída: arquivo de vídeo .avi (ex.: 'saida.avi') no codec XVID.
cap = cv.VideoCapture(0)

# Largura e altura atuais dos quadros (devem casar com o VideoWriter)
width = cap.get(cv.CAP_PROP_FRAME_WIDTH)    # float
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT)  # float
fps = 10.0                                  # taxa adequada à webcam usada

# Codec XVID + objeto VideoWriter
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('saida.avi', fourcc, fps, (int(width), int(height)))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # frame = cv.flip(frame, 0)   # <- LINHA REMOVIDA: causava imagem de cabeça para baixo
    out.write(frame)              # grava o quadro (já na orientação correta)
    cv.imshow('frame', frame)

    if cv.waitKey(1) == ord('q'):
        break

cap.release()
out.release()                    # IMPORTANTE: fecha o arquivo de vídeo corretamente
cv.destroyAllWindows()
````

**Explicação detalhada (pergunta do enunciado):**

- **Problema encontrado originalmente:** no exemplo-base do tutorial do OpenCV existe a linha
  `frame = cv.flip(frame, 0)`, que espelha o quadro no eixo vertical. Com ela ativa, o vídeo
  gravado ficava **de cabeça para baixo**. Além disso, um `fps` incompatível com a captura faz o
  vídeo parecer acelerado ou arrastado na reprodução.
- **Causa:** `cv.flip(frame, 0)` inverte verticalmente cada quadro **antes** de gravá-lo; e o
  parâmetro `fps` do `VideoWriter` é quem determina a velocidade com que o arquivo será
  reproduzido.
- **Solução aplicada:** **remover/comentar** o `cv.flip` (a imagem passa a ser gravada na
  orientação normal) e **ajustar o `fps`** para um valor adequado à webcam utilizada
  (`fps = 10.0`), garantindo reprodução em velocidade natural.
- **Parâmetros do `VideoWriter`:** `('saida.avi', fourcc=XVID, fps=10.0, (width, height))`, com
  `width`/`height` obtidos de `cap.get(cv.CAP_PROP_FRAME_WIDTH/HEIGHT)` para casar exatamente com
  o tamanho dos quadros. Não esquecer de chamar `out.release()` ao final para finalizar o arquivo.

**Resultado/execução:** rodando o programa, a janela mostra a câmera ao vivo e os quadros são
gravados; pressionar **`q`** encerra e finaliza o `.avi`. A equipe usou exatamente esta base
(alterando apenas o nome do arquivo de saída) para gravar todos os vídeos da Parte 2.

---

**Pergunta final do enunciado — "Se for necessário alterar a imagem (algum processamento), em que
ponto dos quatro programas isso deve ser realizado?"**

O processamento deve ser feito **depois de obter o quadro** (`cv.imread` no programa de imagem, ou
`cap.read()` nos programas de vídeo/webcam) e **antes de usá-lo na saída** — isto é, antes de
`cv.imshow` (exibição) e/ou antes de `out.write` (gravação). É exatamente o ponto onde, no programa
(D), está a linha comentada do `cv.flip`: qualquer operação (conversão de cor, filtro, redimensionar,
desenhar sobre o quadro, etc.) entra **entre a captura/leitura e a exibição/gravação**, de modo que o
resultado processado seja o que aparece na janela e/ou é salvo no arquivo.

## 5. Análise e discussão

**Funcionamento dos programas.** Os quatro experimentos cobrem o ciclo entrada → processamento →
saída da visão computacional. O item (A) mostrou que a *flag* de `cv.imread` determina o espaço de
cor da imagem carregada (o `0` resulta em tons de cinza). O item (B) evidenciou que a velocidade de
reprodução é função apenas do **atraso entre quadros**, controlado por `cv.waitKey`, com a relação
`fps ≈ 1000/atraso`. O item (C) demonstrou a captura de teclas em tempo real para salvar um quadro
da webcam, e o item (D) tratou da gravação com `VideoWriter`, com atenção à orientação da imagem e
ao tamanho/`fps` corretos.

**Alterações realizadas (e inconsistências corrigidas neste relatório).**
- *(B)* O comentário do código original ("40ms - 25fps") não correspondia à espera efetiva
  (`time.sleep(1/1000.0)` + `waitKey(1)`, ≈1 ms). O relatório explicita os ajustes corretos para
  velocidade natural (`waitKey(42)`), rápida (`waitKey(1)`) e lenta (`waitKey(100)`).
- *(C)* O script original salvava em `pessoalento.png` e chamava `waitKey(1)` duas vezes; a versão
  do relatório salva em **`foto1.png`** (conforme o enunciado) e lê a tecla **uma única vez** por
  iteração.
- *(D)* Destacou-se que a linha `cv.flip(frame, 0)` é a causa da imagem invertida; sua remoção
  resolve a orientação. Mantido `fps = 10.0` para velocidade adequada.

**Dificuldades encontradas e soluções.** As principais dificuldades foram (i) entender que a
velocidade do vídeo depende do atraso e não do arquivo — resolvida ajustando `waitKey`; (ii) o vídeo
gravado "de cabeça para baixo" — resolvido removendo o `flip`; e (iii) garantir que o tamanho
passado ao `VideoWriter` coincida com o dos quadros capturados — resolvido lendo
`CAP_PROP_FRAME_WIDTH/HEIGHT` diretamente da câmera.

**Compatibilidade com OpenCV moderno.** Todos os comandos usados (`imread`, `imshow`, `waitKey`,
`VideoCapture`, `VideoWriter`, `VideoWriter_fourcc`, `imwrite`) são válidos no OpenCV 4.x. Boa
prática adotada: usar `cv.waitKey(1) & 0xFF` ao comparar com `ord(...)` para portabilidade.
